```python
import os
import cv2
from PIL import Image
from ipywidgets import Video
os.chdir('..')
print(os.getcwd())
from src.predict import YOLOInference
```

```python
def create_video(frames, video_path, fps=20):
    """
    Create a video from a list of frames and save it to the specified path.

    Args:
        frames (list): List of frames (images) to be included in the video.
        video_path (str): Path where the video will be saved.
        fps (int, optional): Frames per second for the video. Defaults to 20.

    Returns:
        None
    """
    # Get the height and width of the frames
    height, width, layers = frames[0].shape
    size = (width, height)
    
    # Define the codec and create the VideoWriter object
    out = cv2.VideoWriter(video_path, cv2.VideoWriter_fourcc(*'VP90'), fps, size)
    
    # Write each frame to the video
    for i in range(len(frames)):
        if i % 7 == 0:
            out.write(frames[i])
    
    # Release the VideoWriter object
    out.release()
```

```python
yolo_inference = YOLOInference(model_path=f'runs\\detect\\train\\weights\\best.pt')
```

## video inference

```python
frame = yolo_inference.predict_video("midia_test\\videos\\v1.mp4",imshow=False)
create_video(frame, f"midia_test\\videos_results\\v1.mp4")
Video.from_file("midia_test\\videos_results\\v1.mp4", width=320, height=320)
```
